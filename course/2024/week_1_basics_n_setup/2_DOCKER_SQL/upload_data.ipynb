```python
import pandas as pd
```

```python
data = pd.read_csv("yellow_tripdata_2021-01.csv.gz",compression='gzip')
print(data.head(10))
```

```python
data
```

```python
print(pd.io.sql.get_schema(data,"yellow_taxi_data"))
```

```python
data.tpep_pickup_datetime = pd.to_datetime(data.tpep_pickup_datetime)
data.tpep_dropoff_datetime = pd.to_datetime(data.tpep_dropoff_datetime)
```

```python
print(pd.io.sql.get_schema(data,"yellow_taxi_data"))
```

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
print(pd.io.sql.get_schema(data, name='yellow_taxi_data', con=engine))
```

```python
df_iter = pd.read_csv("yellow_tripdata_2021-01.csv.gz",compression='gzip', iterator=True, chunksize=100000)
df = next(df_iter)
```

```python
df
```

```python
from time import time
while True: 
    t_start = time()

    df = next(df_iter)

    df.tpep_pickup_datetime = pd.to_datetime(df.tpep_pickup_datetime)
    df.tpep_dropoff_datetime = pd.to_datetime(df.tpep_dropoff_datetime)
    
    df.to_sql(name='yellow_taxi_data', con=engine, if_exists='append')

    t_end = time()

    print('inserted another chunk, took %.3f second' % (t_end - t_start))
```
